* There are more number of outlier in runs_x .
* Need to remove the outliers.

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1

df=df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]
```

```python
df.runs_x=np.sqrt(df.runs_x)
df.no=np.sqrt(df.no)
df.avg_x=np.sqrt(df.avg_x)
df.bf=np.sqrt(df.bf)
df.x100=np.sqrt(df.x100)
df.x50=np.sqrt(df.x50)
df.x4s=np.sqrt(df.x4s)
df.x6s=np.sqrt(df.x6s)
df.wkts=np.sqrt(df.wkts)
df.avg_y=np.sqrt(df.avg_y)
df.sr_y=np.sqrt(df.sr_y)
df.x4w=np.sqrt(df.x4w)
df.x5w=np.sqrt(df.x5w)
```

```python
plt.figure(figsize=(20,18))
sns.boxplot(data=df)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df.iloc[:,:]=ss.fit_transform(df.iloc[:,:])
df.head(2)
```

```python
from sklearn.cluster import KMeans
```

```python
interclusterd=[]
for i in range(1,9):
    k=KMeans(n_clusters=i)
    k.fit(df)
    
    interclusterd.append(k.inertia_)
```

```python
interclusterd
```

```python
plt.figure()
plt.plot(range(1,9),interclusterd,'*--')
plt.xlabel('K')
plt.ylabel('WCD')
plt.show()
```

With the elbow method we can conclude from the above visual representation that 3 clusters would be efficient for our data.

```python
k=KMeans(n_clusters=3)
k.fit(df)
```

```python
k.labels_
```

```python
df['labels']=k.labels_
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
pc=PCA(n_components=3,random_state=1)
components=pc.fit_transform(df)
```

```python
pc1=components[:,0]
pc2=components[:,1]
pc3=components[:,2]
```

```python
ax=plt.axes(projection='3d')
ax.scatter(pc1,pc2,pc3,c=df['labels'])
plt.show()
```

Above visualisation shows that our data seems to merge whish isnt good as there should be high intracluster distance between groups.

```python
df['labels'].value_counts()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1,test_size=0.2,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression 
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
ac=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
print(f'Accuracy:{ac}\n {cm} \n{cr}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
ac=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
print(f'Accuracy:{ac}\n {cm} \n{cr}')
```

Using Logistic Regressor Accuracy is 95% and Using KNN Classifier Accuracy is 100%.
